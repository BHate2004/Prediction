```python
import pandas as pd
import numpy as np
```

```python
# pd.set_option('display.max_rows',None)
# pd.set_option('display.max_columns',None)
```

```python

df=pd.read_csv("/content/flats.csv"
)
df.head(2)
new_df=pd.read_csv("/content/houses.csv")
```

```python
df["price"].shape
```

```python

df.rename(columns={"area":"area_sqm"},inplace=True)

new_df.rename(columns={"rate":"area_sqm"},inplace=True)
```

```python
df["society"]=df["society"].str.replace('s*\d+(\.\d+)?\s*★?',"",regex=True)
new_df["society"]=new_df["society"].str.replace('s*\d+(\.\d+)?\s*★?',"",regex=True)
```

```python

df=df[~df["price"].str.contains("on",na=False)]
new_df=new_df[~new_df["price"].str.contains("on",na=False)]
new_df.head(2)
```

```python
df["price"].str.split(" ").str[0].value_counts()
```

```python
df["price"].value_counts()
```

```python
df["price"].fillna(0,inplace=True)

df["code"]=df["price"].str.split(" ").str[1]
df["price"]=df["price"].str.split(" ").str[0].astype("float")
new_df["price"].fillna(0,inplace=True)

new_df["code"]=new_df["price"].str.split(" ").str[1]
new_df["price"]=new_df["price"].str.split(" ").str[0].astype("float")

```

```python
def convert_price(row):
  if row["code"]=="Lac":
    return (row["price"]/100)
  elif row["code"]=="Crore":
      return (row["price"])
  else:
    return row["price"]
```

```python
df["price"]=df.apply(convert_price,axis=1)
new_df["price"]=new_df.apply(convert_price,axis=1)
```

```python
new=df["area_sqm"].str.split(" ").str[1].str.split("/")
df["area_sqm"]=new.str[0]
df["area_sqm"]=df["area_sqm"].str.replace(",","").astype("float")

```

```python
new_df["area_sqm"]=new_df["area_sqm"].str.extract(r"(\d+)").astype("float")
new_df["area"]=new_df["area"].str.extract(r"(\d+)").astype("float").astype("float")
new_df.rename(columns={"area":"Area"},inplace=True)
```

```python
df["bedRoom"].value_counts()
new_df["bedRoom"].value_counts()
```

```python
df=df[~df["bedRoom"].isnull()]
df["bedRoom"]=df["bedRoom"].str.strip(" ").str[0].astype("int")
new_df=new_df[~new_df["bedRoom"].isnull()]
```

```python
df["bathroom"]=df["bathroom"].str.split(" ").str[0].astype("int")
new_df["bathroom"]=new_df["bathroom"].str.split(" ").str[0].astype("int")
```

```python
df["balcony"]=df["balcony"].str.split(" ").str[0].replace("No",0)
new_df["balcony"]=new_df["balcony"].str.split(" ").str[0].replace("No",0)
```

```python
df["additionalRoom"].fillna("not_available",inplace=True)
new_df["additionalRoom"].fillna("not_available",inplace=True)

```

```python
def floor(x):
  if x=="Ground" or  x=="Lower":
    return 0
  elif x=="Basement":
   return -1
  else:
    return x
```

```python
df["floorNum"]=df["floorNum"].str.strip(r"^s+").str.split(" ").str[0].apply(floor)
new_df["noOfFloor"]=new_df["noOfFloor"].str.strip(r"^s+").str.split(" ").str[0].apply(floor)
new_df.rename(columns={"noOfFloor":"floor_Num"})
```

```python
df['Floor_no']=df["floorNum"].str.strip(r"^s+").str.split(" ").str[0].str.extract(r"(\d+)")
```

```python
df["facing"].fillna("N/A",inplace=True)
new_df["facing"].fillna("N/A",inplace=True)
df.head(2)

```

```python
df.drop(columns="link",inplace=True)
df.drop(columns="floorNum",inplace=True)
new_df.drop(columns="link",inplace=True)
```

```python
df.rename(columns={"area_sqm":"price_per_sqft"},inplace=True)
new_df.rename(columns={"area_sqm":"price_per_sqft"},inplace=True)
```

```python

df.insert(loc=4,column="Area",value=round((df["price"]*10000000)/df["price_per_sqft"]))
df["property_type"]="Flat"
new_df["property_type"]="Houses"

```

```python
df.drop(columns="code",inplace=True)
new_df.drop(columns="code",inplace=True)
```

```python

new_df.rename(columns={"noOfFloor":"Floor_no"},inplace=True)
new_df.head(2)
```

```python
universal=pd.concat([df,new_df],axis=0)
universal.drop(columns="property_id",inplace=True)
```
